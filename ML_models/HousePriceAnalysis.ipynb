```python
df=data
print(df.shape)
def remove_outliers(df):
    for column in df.columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df['z_score']=(df[column]-df[column].mean())/df[column].std()
        df = df[((df[column] >= lower_bound) & (df[column] <= upper_bound)) & ((df['z_score'] >= -3.0) & (df['z_score'] <= 3.0))]
    df.drop(columns=['z_score'],inplace=True)
    return df
cleaned_data=remove_outliers(df)
print(cleaned_data.shape)
```

```python
plt.figure(figsize=(4,4))
for column in cleaned_data.columns:
    plt.figure()  # Create a new figure for each box plot
    cleaned_data.boxplot(column)  # Create a box plot for the current column
    plt.title('Box plot of ' + column)  # Set the title of the box plot
    plt.show() 
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
X_train, X_test, y_train, y_test=train_test_split(cleaned_data.drop(columns=['price']),cleaned_data['price'],test_size=0.2,random_state=40)
X_train.shape,X_test.shape
```

```python
linear_regression_model= Pipeline([('scaler',StandardScaler()),('model',LinearRegression())])
linear_regression_model.fit(X_train,y_train)
```

```python
ypred=linear_regression_model.predict(X_test)
```

```python
linear_regression_model.score(X_test,y_test)
```

```python
linear_regression_model.score(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
regressor.score(X_test,y_test)
```
