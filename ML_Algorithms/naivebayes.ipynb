```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_wine 
```

```python
wine=load_wine()
```

```python
X_train,X_test,y_train,y_test=train_test_split(wine.data,wine.target,test_size=0.3,random_state=30)
```

```python
gnb=GaussianNB()
gnb.fit(X_train,y_train)
```

```python
gnb.predict(X_test)
gnb.score(X_test,y_test)
```

```python
gnb.predict_proba(X_test)
```

### MultinomialNB

```python
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data=pd.read_csv(r'C:\Users\pramo\Datasets\datasets\emails.csv')
data
```

```python
data['spam']=data['Category'].apply(lambda x:1 if x=='spam' else 0)
```

```python
data.drop(columns=['Category'],inplace=True)
```

```python
data.rename(columns={"Message":"text"},inplace=True)
```

```python
X_train,X_test,y_train,y_test=train_test_split(data['text'],data['spam'],test_size=0.2,random_state=30)
X_train,y_train
cv=CountVectorizer()
X_train_t=cv.fit_transform(X_train)
X_test_t=cv.transform(X_test)
#X_train_t.toarray()
#list=list(cv.get_feature_names_out())
```

```python
mnb=MultinomialNB()
mnb.fit(X_train_t,y_train)
```

```python
mnb.predict(X_test_t)
mnb.score(X_test_t,y_test)
```

```python
mnb.predict(cv.transform(['Subject: naturally irresistible your corporate...']))
```

```python
from sklearn.pipeline import Pipeline
model=Pipeline([('vectorizer',CountVectorizer()), 
               ('nvb',MultinomialNB())
               ])
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```
