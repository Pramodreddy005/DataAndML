## KNN

### Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
from sklearn.datasets import load_digits
```

```python
digit=load_digits()
```

```python
X_train,X_test,y_train,y_test=train_test_split(digit.data,digit.target,test_size=0.3,random_state=40)
X_train.shape,X_test.shape
```

```python
knn=KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
knn.score(X_test,y_test)
```

```python
c=confusion_matrix(y_test,y_pred)
disp=ConfusionMatrixDisplay(c,display_labels=knn.classes_)
disp.plot()
```

### Regression

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import MinMaxScaler
```

```python
housing=fetch_california_housing()
housing.data
scaler=MinMaxScaler()
scaled_data=scaler.fit_transform(housing.data)
scaled_target=scaler.fit_transform((housing.target.reshape(-1,1)))
```

```python
X_train,X_test,y_train,y_test=train_test_split(scaled_data,scaled_target,test_size=0.3,random_state=42)
X_train,y_train,y_train,y_test
```

```python
knn_reg=KNeighborsRegressor(n_neighbors=10)
knn_reg.fit(X_train,y_train)
```

```python
y_pred=knn_reg.predict(X_test)
knn_reg.score(X_test,y_test)
```
