## Support Vector Machine

### Classification

```python
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,precision_score,accuracy_score,recall_score,roc_auc_score
```

```python
iris_data=load_iris()
```

```python
iris_data.data,iris_data.feature_names,iris_data.target,iris_data.target_names
```

```python
X_train,X_test,y_train,y_test=train_test_split(iris_data.data,iris_data.target,test_size=0.3,random_state=30)
svm_model=SVC(C=2,gamma=5)
svm_model.fit(X_train,y_train)
```

```python
svm_model.score(X_test,y_test),svm_model.score(X_train,y_train)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
y_pred=svm_model.predict(X_test)
conf_matrix = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=['setosa', 'versicolor', 'virginica'])
disp.plot(cmap='viridis',values_format='d')
```
