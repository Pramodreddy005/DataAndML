```python
import pandas as pd
from sklearn.datasets import fetch_california_housing

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

```python
data=fetch_california_housing()
```

```python
df = pd.DataFrame(data=data.data, columns=data.feature_names)
df['ashtarget_name'] = data.target
df.to_csv(r"C:\Users\pramo\Datasets\datasets\sklearncsv\fetch_california_housing.csv", index=False)
```

```python
data1=pd.read_csv(r"C:/Users/pramo/Datasets/datasets/sklearncsv/fetch_california_housing.csv")
data1.rename(columns={'ashtarget_name':'median_price'},inplace=True)
```

```python
data1.columns
X_train,X_test,y_train,y_test= train_test_split(data1.drop(columns=['median_price']),data1['median_price'],test_size=0.2)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
msr=mean_absolute_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
rscore=r2_score(y_test,y_pred)
print(msr,mae,rscore)
```

### Logistic Regression

#### Binary Classification

```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score
```

```python
cancer_data=load_breast_cancer()
print(cancer_data.feature_names)
X_train,X_test,y_train,y_test=train_test_split(cancer_data.data,cancer_data.target,test_size=0.3)
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
logistic_regression_model=LogisticRegression(max_iter=300)
logistic_regression_model.fit(X_train_scaled,y_train)
```

```python
y_pred=logistic_regression_model.predict(X_test_scaled)
a_score=accuracy_score(y_test,y_pred)
p_score=precision_score(y_test,y_pred)
r_score=recall_score(y_test,y_pred)
f_score=f1_score(y_test,y_pred)
#ruc_auc=ruc_auc_score(y_test,y_pred_proba([)

print(a_score, p_score, r_score, f_score, sep='\n \n')
```

### Multiclass LogisticRegression

```python
from sklearn.datasets import load_wine
```

```python
wine_data=load_wine()
X_train,X_test,y_train,y_test=train_test_split(wine_data.data,wine_data.target,test_size=0.3,random_state=30)
#wine_data.data 
```

```python
multi_log_reg=LogisticRegression(max_iter=200,multi_class='ovr')
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
multi_log_reg.fit(X_train_scaled,y_train)
```

```python
y_pred=multi_log_reg.predict(X_test_scaled)
```

```python
a_score=accuracy_score(y_test,y_pred)
p_score=precision_score(y_test,y_pred,average='macro')
r_score=recall_score(y_test,y_pred,average='macro')
f_score=f1_score(y_test,y_pred,average='macro')
print(a_score,p_score,r_score,f_score,sep='\n')
y_p_p=multi_log_reg.predict_proba(X_test_scaled)
r=roc_auc_score(y_test,y_p_p,average='macro',multi_class='ovr')
#y_p_p
```
